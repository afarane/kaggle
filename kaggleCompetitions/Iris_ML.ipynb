# Data Visualization

# Try Model Predictions 

Classify Data

```python
X = iris_data.drop(['Id','Species'],axis=1)
y = iris_data['Species']
print(X.shape)
print(y.shape)
```

LogisticRegression

```python
logreg = LogisticRegression()
logreg.fit(X,y)
y_pred = logreg.predict(X)
print(metrics.accuracy_score(y,y_pred))
```

KNN Classifier

```python
knn = KNeighborsClassifier(n_neighbors=12)
knn.fit(X,y)
y_pred = knn.predict(X)
print(metrics.accuracy_score(y,y_pred))
```
